**Import Libraries**

```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
```

**Load HuggingFace Mode**l

```python
generator = pipeline(
    "text-generation",
    model="google/flan-t5-large"
)
```

**Load dataset**

```python
df = pd.read_csv("dataset.csv")

print(df.head())
```

**Zero Shot**

```python
def zero_shot(q):
    return f"""
Solve this problem.

Question: {q}
Answer:
"""
```

**Few Shot**

```python
def few_shot(q):
    return f"""
Example:
Q: What is 10% of 200?
A: 20

Example:
Q: If speed is 60 km/h for 2 hours?
A: 120

Now solve:
Q: {q}
A:
"""
```

**COT(Chain of thoughts)**

```python
def cot_prompt(q):
    return f"""
Question: {q}

Let's think step by step.
Give final answer at end.
"""
```

**Interview Prompt**

```python
def interview_prompt(q):
    return f"""
Question: {q}

1. What is question asking?
2. What formula or logic is needed?
3. Solve step by step.
4. Give final answer.
"""
```

**ToT(Tree of Thoughts)**

```python
def tot_prompt(q):
    return f"""
Question: {q}

Give 3 different solution approaches.
Evaluate each approach.
Choose best approach and give final answer.
"""
```

**Model Run Function**

```python
def run_model(prompt):
    output = generator(prompt, max_length=200)
    return output[0]["generated_text"]
```

**Run All Prompt Methods On Dataset**

```python
results = []

for _, row in tqdm(df.iterrows(), total=len(df)):

    q = row["question"]
    true_ans = row["answer"]

    res = {
        "question": q,
        "true_answer": true_ans,
        "zero_shot": run_model(zero_shot(q)),
        "few_shot": run_model(few_shot(q)),
        "cot": run_model(cot_prompt(q)),
        "interview": run_model(interview_prompt(q)),
        "tot": run_model(tot_prompt(q))
    }

    results.append(res)
```

**Convert Results To DataFrame**

```python
results_df = pd.DataFrame(results)

print(results_df.head())
```

**Saved result**

```python
results_df.to_csv("prompt_engineering_results.csv", index=False)

print("Results saved successfully!")
```

**Accuracy Checking**

```python
def check_correct(pred, true):
    return str(true) in str(pred)

for col in ["zero_shot","few_shot","cot","interview","tot"]:
    results_df[col+"_correct"] = results_df.apply(
        lambda x: check_correct(x[col], x["true_answer"]), axis=1
    )

print(results_df.mean(numeric_only=True))
```

**Experimental results show that Chain-of-Thought prompting achieved the highest accuracy (53%), followed by Interview prompting (50%). Few-shot prompting showed moderate improvement over zero-shot prompting. Tree-of-Thought prompting showed lower performance due to model limitations in evaluating multiple reasoning branches. These results indicate that structured reasoning prompts significantly improve performance in reasoning tasks.**
