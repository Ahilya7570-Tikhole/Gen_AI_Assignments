**LOAD A TACOTRON-LINEAGE MODEL**

```python
from transformers import pipeline
import soundfile as sf
```

**GENERATE SPEECH**

```python
tts = pipeline(
    "text-to-speech",
    model="suno/bark-small", # Changed to a recognized Hugging Face text-to-speech model
    device=0
)
```

**PLAY AUDIO**

```python
speech = tts("Isha and firdos are watching blinkit offers")
# Squeeze the audio array to a 1D array for soundfile (removes dimensions of size 1)
audio_data = speech["audio"].squeeze()
sf.write("output.wav", audio_data, speech["sampling_rate"], format='WAV', subtype='FLOAT')
```
