**Import Required Libraries**

```python
from huggingface_hub import InferenceClient
from PIL import Image
import matplotlib.pyplot as plt
```

**Huggingface Token**

```python
HF_TOKEN = "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"
```

**Create Inference Client**Hugging Face provides methods like text_to_image() to directly generate images from text prompts.

```python
client = InferenceClient(token=HF_TOKEN)
```

**Text Input,Generate Image,Display Image**

```python
prompt = input("Enter image description: ")
image = client.text_to_image(
    prompt,
    model="stabilityai/stable-diffusion-xl-base-1.0"
)
plt.imshow(image)
plt.axis("off")
plt.show()
```

**Save the Image**

```python
image.save("generated_image.png")
print("Image saved successfully")
```
