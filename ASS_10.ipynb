**Import libraries and stable diffusion model**

```python
import torch
from diffusers import StableDiffusionPipeline
```

**Load Pretrained Stable Diffusion Model**

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

**Take Prompt Input From User and save image**

```python
prompt = input("Enter your image prompt: ")

image = pipe(prompt).images[0]

image.save("generated_image.png")

image
```

```python
image = pipe(
    prompt,
    guidance_scale=7.5,
    num_inference_steps=40
).images[0]
```

**Assignment demonstrates multimodal transformer capability by using Stable Diffusion to convert natural language text into images. The model uses transformer-based attention to understand important words in the prompt and diffusion to generate images. Since the system processes text and produces visual output using shared embeddings, it proves multimodal learning and cross-modal understanding.**
